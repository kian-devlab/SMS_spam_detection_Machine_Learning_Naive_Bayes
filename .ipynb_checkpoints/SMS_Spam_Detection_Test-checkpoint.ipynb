```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import nltk
import string
import pprint
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
from collections import Counter
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import cross_val_score
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import roc_curve, auc
%matplotlib inline
```

```python
df_sms = pd.read_csv('spam.csv',encoding='latin-1')
df_sms.head()
```

```python
df_sms = df_sms.drop(["Unnamed: 2", "Unnamed: 3", "Unnamed: 4"], axis=1)
df_sms = df_sms.rename(columns={"v1":"label", "v2":"sms"})
df_sms.head()
```

```python
print(len(df_sms))
```

```python
df_sms.label.value_counts()
```

```python
df_sms.describe()
```

```python
df_sms['length'] = df_sms['sms'].apply(len)
df_sms.head()
```

```python
df_sms['length'].plot(bins=50, kind='hist')
```

```python
df_sms.hist(column='length', by='label', bins=50,figsize=(10,4))
```

```python
df_sms.loc[:,'label'] = df_sms.label.map({'ham':0, 'spam':1})
print(df_sms.shape)
df_sms.head()
```

```python
documents = ['Hello, how are you!',
             'Win money, win from home.',
             'Call me now.',
             'Hello, Call hello you tomorrow?']

lower_case_documents = []
lower_case_documents = [d.lower() for d in documents]
print(lower_case_documents)
```

```python
sans_punctuation_documents = []
for i in lower_case_documents:
    sans_punctuation_documents.append(i.translate(str.maketrans("","", string.punctuation)))

sans_punctuation_documents
```

```python
preprocessed_documents = [[w for w in d.split()] for d in sans_punctuation_documents]
preprocessed_documents
```

```python
frequency_list = []
frequency_list = [Counter(d) for d in preprocessed_documents]
pprint.pprint(frequency_list)
```

```python
count_vector = CountVectorizer()
```

```python
count_vector.fit(documents)
count_vector.get_feature_names_out()
```

```python
doc_array = count_vector.transform(documents).toarray()
doc_array
```

```python
frequency_matrix = pd.DataFrame(doc_array, columns = count_vector.get_feature_names_out())
frequency_matrix
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_sms['sms'],
                                                    df_sms['label'],test_size=0.20,
                                                    random_state=1)
```

```python
# Instantiate the CountVectorizer method
count_vector = CountVectorizer()

# Fit the training data and then return the matrix
training_data = count_vector.fit_transform(X_train)

# Transform testing data and return the matrix.
testing_data = count_vector.transform(X_test)
```

```python
y_train = np.array(y_train, dtype=int)
```

```python
naive_bayes = MultinomialNB()
naive_bayes.fit(training_data, y_train)
```

```python
naive_bayes.fit(training_data, y_train)
```

```python
predictions = naive_bayes.predict(testing_data.toarray())
```

```python
y_test = y_test.astype(int)  # or float, depending on your data
predictions = predictions.astype(int)  # or float
```

```python
print("Unique values in y_test:", np.unique(y_test))
print("Unique values in predictions:", np.unique(predictions))
```

```python
print("Shape of y_test:", y_test.shape)
print("Shape of predictions:", predictions.shape)
```

```python
print('Accuracy score: {}'.format(accuracy_score(y_test, predictions)))
print('Precision score: {}'.format(precision_score(y_test, predictions)))
print('Recall score: {}'.format(recall_score(y_test, predictions)))
print('F1 score: {}'.format(f1_score(y_test, predictions)))
```

```python
# Define the metrics
metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
values = [accuracy_score(y_test, predictions),
          precision_score(y_test, predictions),
          recall_score(y_test, predictions),
          f1_score(y_test, predictions)]

# Create a bar plot
plt.figure(figsize=(8, 6))
plt.bar(metrics, values, color=['blue', 'orange', 'green', 'purple'])
plt.ylim(0, 1)  # Set y-axis limits to [0, 1]
plt.title('Model Performance Metrics')
plt.ylabel('Score')
plt.grid(axis='y')

# Annotate the bars with their values
for i, v in enumerate(values):
    plt.text(i, v + 0.02, f'{v:.2f}', ha='center', va='bottom')

plt.show()
```
